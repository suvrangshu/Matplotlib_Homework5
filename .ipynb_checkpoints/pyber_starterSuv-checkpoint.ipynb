```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

#-------------------------
#Author : Suvrangshu Ghosh
#--------------------------
#Modification history:
#---------------------
#
#
#
#---------------------------

# File to Load (Remember to change these)
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"

# Read the City and Ride Data
city = pd.read_csv(city_data_to_load)
ride = pd.read_csv(ride_data_to_load)

# Combine the data into a single dataset
combdata = pd.merge(ride,city,on="city")

# Display the data table for preview
combdata.to_csv("data/city_data1.csv", index=False, header=True)
#print(combdata)
```

## Bubble Plot of Ride Sharing Data

```python
#Getting specific data from combdata

urban = combdata.loc[combdata["type"] == "Urban"] # type eq Urban
suburban = combdata.loc[combdata["type"] == "Suburban"] # type eq suburban
rural = combdata.loc[combdata["type"] == "Rural"] # type eq rural

#Urban
urban_city_group = urban.groupby("city")
urban_avg_fare = urban_city_group.fare.mean()
urban_tot_rides = urban_city_group.date.count()
urban_tot_drivers = urban_city_group.driver_count.sum()

urban = plt.scatter(
        urban_tot_rides, 
        urban_avg_fare, 
        marker="o", 
        c="lightcoral", 
        edgecolors="black",
        s=urban_tot_drivers**0.85,
        linewidths=0.85,
        alpha = 1)

#suburban
suburban_city_group = suburban.groupby("city")
suburban_avg_fare = suburban_city_group.fare.mean()
suburban_tot_rides = suburban_city_group.date.count()
suburban_tot_drivers = suburban_city_group.driver_count.sum()

sub = plt.scatter(
        suburban_tot_rides, 
        suburban_avg_fare, 
        marker="o", 
        c="lightskyblue", 
        edgecolors="black",
        s=suburban_tot_drivers**0.85,
        linewidths=0.85,alpha = 1)


#suburban
rural_city_group = rural.groupby("city")
rural_avg_fare = rural_city_group.fare.mean()
rural_tot_rides = rural_city_group.date.count()
rural_tot_drivers = rural_city_group.driver_count.sum()

rural = plt.scatter(
        rural_tot_rides, 
        rural_avg_fare, 
        marker="o", 
        c="gold", 
        edgecolors="black",
        s=rural_tot_drivers**0.85,
        linewidths=0.85,
        alpha = 1)

# Obtain the x and y coordinates for each of the three city types
plt.grid()
# Build the scatter plots for each city types
plt.xlabel("Total Number of Rides (Per City)")
plt.ylabel("Average Fare ($)")
plt.title("Pyber Ride Sharing Data (2016)")

lgnd = plt.legend(
        (urban,sub,rural),
        ("Urban","Suburban","Rural"),
        fontsize='small',
        loc="best",
        title = "City Types",
        numpoints=2,
        markerscale=0.5,
        scatterpoints=1)

plt.text(42,35,"Note: \nCircle size correlates with driver count per city")
lgnd.legendHandles[0]._sizes = [30]
lgnd.legendHandles[1]._sizes = [30]
lgnd.legendHandles[2]._sizes = [30]

# Save the plot and display it
plt.savefig("Images/PyberRideShareData2016.png")

plt.show()
# Incorporate the other graph properties

# Create a legend

# Incorporate a text label regarding circle size

# Save Figure
#plt.scatter(total_rides, average_fare, marker="o", c=["Gold","LightSkyBlue","LightCoral"], edgecolors="black",
#            s=total_drivers**0.85,linewidths=0.85,alpha = 1)
```

## Total Fares by City Type

```python
# Calculate Type Percents
city_type = combdata.groupby("type")
total = city_type.fare.sum().sum()
total_fare = (city_type.fare.sum()/total)*100

#print(total_fare.index)
#
#urban_percent
plt.pie(
    total_fare,labels =total_fare.index, 
    colors=["Gold", "lightskyblue","lightcoral"],
    autopct="%1.1f%%",
    startangle=150,
    shadow = True,
    wedgeprops   = { 'linewidth' : 2,
                     'edgecolor' : "black" },
    explode = (0, 0, 0.1))

plt.title("% of Total Fare by City Type")

#size of plot
fig = plt.gcf()
fig.set_size_inches(5,5) #size of plot

plt.savefig("Images/PieChart_TotalFare_by_City_type%.png")
# Build Pie Chart
#autopct =urban_percent.map("{:.1f}%".format),
#urban_percent = urban_percent.map("{:.1f}%".format)
# Save Figure
```

## Total Rides by City Type

```python
# Calculate Ride Percents
city_type = combdata.groupby("type")

total = city_type.ride_id.count().sum()
total_fare = (city_type.ride_id.count()/total)*100

# Build Pie Chart
plt.pie(
    total_fare,labels =total_fare.index, 
    colors=["Gold", "lightskyblue","lightcoral"],
    autopct="%1.1f%%",
    startangle=150,
    shadow = True,
    wedgeprops   = { 'linewidth' : 2,
                     'edgecolor' : "black" },
    explode = (0, 0, 0.1))

plt.title("% of Total Rides by City Type")


#size of plot to make circle big/small
fig = plt.gcf()
fig.set_size_inches(5,5) #size of plot

# Save Figure
plt.savefig("Images/PieChart_Percent_of_TotalRides_by_City_type.png")


```

```python
# Show Figure
plt.show()
```
